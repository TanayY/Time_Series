```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

## Import MetRail Data

```python
data = pd.read_csv('/content/drive/MyDrive/Datasets/MSDS/Course 6/Time Series Forecasting/Metrail_Dataset/Metrail_dataset.csv')
data.columns=['Month', 'Travellers']
data['Month'] = pd.to_datetime(data['Month'], format='%b-%y')
data = data.set_index('Month')
```

```python
data.head()
```

## Plot time series data

```python
data.plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('MetRail Ridership Time Series Plot')
plt.show(block=False)
```

```python
data = data.assign(Travellers_Last_Observed = data.Travellers.fillna(method='pad'))
data[['Travellers_Last_Observed']].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Missing value: Last Observed Imputation')
plt.show(block = False)
```

```python
data['Travellers_Last_Observed'].mean()
```

```python
data = data.assign(Travellers_Next_Observed = data.Travellers.fillna(method='bfill'))
data[['Travellers_Next_Observed']].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Missing value: Next Observed Imputation')
plt.show(block = False)
```

```python
data['Travellers_Next_Observed'].mean()
```

```python
data = data.assign(Travellers_Linear_Interpolation = data.Travellers.interpolate(method = 'linear'))
data[['Travellers_Linear_Interpolation']].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Missing value: Linear Interpolation')
plt.show(block = False)
```

```python
data['Travellers_Linear_Interpolation'].mean()
```
